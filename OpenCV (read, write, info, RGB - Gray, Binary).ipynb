## Converted to RGB image into Gray Scale image

```python
img2 = cv2.imread('C:/Users/Mehedi Hassan Galib/Desktop/Python/datas/dog.31.jpg')
gray_img = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

cv2.imshow('Gray Scale Image:', gray_img)

#Save the new image
cv2.imwrite('C:/Users/Mehedi Hassan Galib/Desktop/Python/datas/gray_scale.jpg', gray_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## RGB into Binary
<br/>

**Note:** The pixel values of a binary image will be only 0 and 1. 0 for black and 1 for white.

```python
#Converted to gray scale
img2 = cv2.imread('C:/Users/Mehedi Hassan Galib/Desktop/Python/datas/dog.31.jpg', 0)

#Converted to binary
ret, bw = cv2.threshold(img2, 127, 255, cv2.THRESH_BINARY)

#Save the binary image
cv2.imwrite('C:/Users/Mehedi Hassan Galib/Desktop/Python/datas/Binary Image.jpg', bw)
```
